```python
import pandas as pd
from vascx.fundus.feature_sets.full_v2 import fs_full_v2
from vascx.fundus.feature_sets.sparsity import fs_sparsity
from vascx.fundus.retina import Retina
```

```python
# Generate display names mapping
full_v2_display_names = Retina.make_feature_display_names(fs_full_v2)
sparsity_display_names = Retina.make_feature_display_names(fs_sparsity)
```

```python
display_names = {**full_v2_display_names, **sparsity_display_names}
```

```python
import json
from pathlib import Path

# Configure output path
output_path = Path("/mnt/oogergo/eyened/models/releases/vascx/11_25/display_names.json")

# Save display names to JSON
with open(output_path, 'w') as f:
    json.dump(display_names, f, indent=2)

print(f"Display names saved to {output_path}")
```

```python
# Display as DataFrame for better readability
df = pd.DataFrame(list(display_names.items()), columns=["Feature Key", "Display Name"])
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
df.sort_values("Feature Key")
```
