```python
from pathlib import Path

from rtnls_enface.utils.plotting import plot_gridfns

from vascx import RetinaLoader
from vascx.features.vascular_densities import VascularDensity
```

```python
ds_path = Path("../../../samples/fundus/ergo/good")
loader = RetinaLoader.from_folder(ds_path, fundus_subfolder="preprocessed_rgb")
```

```python
r = loader[0]
```

```python
def make_plot_fn(retina):
    def plot_fn(ax, fig):
        VascularDensity().plot(retina.arteries, ax=ax)

    return plot_fn
```

```python
plot_gridfns([make_plot_fn(r) for r in loader[0:4]], ncols=2)
```
