```python
from pathlib import Path

from vascx.fundus.loader import RetinaLoader
from vascx.utils.analysis import (
    extract_in_parallel,
)
```

## Feature extraction

This notebooks runs feature extraction using the output of the models ran in [1_segment_preprocessed.ipynb](1_segment_preprocessed.ipynb)
1. The `RetinaLoader` utility takes care of data loading.
2. VascX runs on the samples in parallel on the CPU
3. We get back a dataframe with an image ID per row and a biomarker per column.

```python
ds_path = Path("../samples/fundus")
```

```python
loader = RetinaLoader.from_folder(ds_path)
```

```python
# adjust n_jobs to your CPU. Recommendation: n = # threads / 2
res = extract_in_parallel(loader.to_dict(), "full_v3", n_jobs=8, print_stack_trace=True)
```

```python
res
```

```python
res.to_csv(ds_path / "biomarkers.csv")
```
