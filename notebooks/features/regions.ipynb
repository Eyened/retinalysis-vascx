```python
from pathlib import Path

import numpy as np
from matplotlib import pyplot as plt
from vascx.fundus import RetinaLoader
from vascx.fundus.features.bifurcation_angles import BifurcationAngles
from vascx.fundus.features.bifurcation_counts import BifurcationCount
from vascx.fundus.features.caliber import Caliber
from vascx.fundus.features.length import Length
from vascx.fundus.features.tortuosity import Tortuosity
from vascx.fundus.features.vascular_densities import VascularDensity
from vascx.fundus.layer import RegionOutOfBoundsError

from rtnls_enface.grids.etdrs import Quadrant
from rtnls_enface.utils.plotting import plot_gridfns
```

```python
ds_path = Path("../../samples/fundus")
loader = RetinaLoader.from_folder(ds_path)
```

```python



def make_plot_fn(feature, retina):
    def plot_fn(ax, fig):
        try:
            
            feature.plot(ax=ax, layer=retina.arteries)
            value = feature.compute(retina.arteries)
            if isinstance(value, dict):
                ax.text(10, 20, str(value), color='w')
            else:
                ax.text(10, 20, f'{np.mean(value):.2f}', color='w')
            
        except RegionOutOfBoundsError as ex:
            print(ex)
            ax.text(10, 20, 'NA', color='w')
            return

    return plot_fn
```

```python
feature = BifurcationAngles(grid_field=Quadrant.Superior)
plot_gridfns([make_plot_fn(feature, r) for r in loader[0:4]], ncols=2)
```

```python
feature = BifurcationCount(grid_field=Quadrant.Inferior)
plot_gridfns([make_plot_fn(feature, r) for r in loader[0:4]], ncols=2)
```

```python
feature = Caliber(grid_field=Quadrant.Temporal)
plot_gridfns([make_plot_fn(feature, r) for r in loader[0:4]], ncols=2)
```

```python
feature = Tortuosity(grid_field=Quadrant.Nasal)
plot_gridfns([make_plot_fn(feature, r) for r in loader[0:4]], ncols=2)
```

```python
feature = VascularDensity(grid_field=Quadrant.Nasal)
plot_gridfns([make_plot_fn(feature, r) for r in loader[0:4]], ncols=2)
```
