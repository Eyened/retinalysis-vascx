## Retipy calibers

Plot the retipy-type calculations

```python
def calc_diameter(segment):
    measurements = segment.calc_diameters_retipy()
    diameters = [m.diameter for m in measurements]
    return np.median(diameters)
```

```python
r.veins.vessels.plot(text=lambda x: f"{calc_diameter(x):.2f}")
```
