```python
device = torch.device("cuda:0")
```

```python
lwnet_outputs_path = "/home/jose/furnace/lwnet/pred"
images_path = "/mnt/ssd1/jose/sinergia/datasets/av_october/images/"
discs_path = "/mnt/ssd1/jose/sinergia/datasets/av_october/discs/"
labels_path = "/mnt/ssd1/jose/sinergia/datasets/av_october/labels/"
loader_lwnet = RetinaLoader.from_folders(
    av_folder=lwnet_outputs_path, fundus_folder=images_path, discs_folder=discs_path
)
loader_labels = RetinaLoader.from_folders(
    av_folder=labels_path, fundus_folder=images_path, discs_folder=discs_path
)
```
