```python
from pathlib import Path

from vascx.fundus import RetinaLoader
from vascx.fundus.features.vascular_densities import VascularDensity

from rtnls_enface.utils.plotting import plot_gridfns
```

```python
ds_path = Path("../../samples/fundus")
loader = RetinaLoader.from_folder(ds_path, fundus_subfolder="preprocessed_rgb")
```

```python
r = loader[0]
```

```python
def make_plot_fn(retina):
    def plot_fn(ax, fig):
        VascularDensity().plot(retina.arteries, ax=ax)

    return plot_fn
```

```python
VascularDensity().plot(r.arteries)
```

```python
plot_gridfns([make_plot_fn(r) for r in loader[0:4]], ncols=2)
```

```python
plot_gridfns([make_plot_fn(r) for r in loader[4:8]], ncols=2)
```
